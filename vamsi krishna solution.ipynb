```python
df2.plot(x ='State/Territory', y='Employment Distribution (%)', kind = 'bar')
plt.ylabel('% of total employee', fontsize=14)
plt.xlabel('states', fontsize=14)#bar chart
```

```python
#line chart for all types of employees in each state 
```

```python
plt.plot(df2['Employment by Occupation - Total'], label='Employment by Occupation - Total', color='green')
plt.plot(df2['Employed Full-Time'], label='Employed Full-Time', color='yellow')

plt.plot(df2['Employed Part-Time'], label='Employed Part-Time', color='purple')

plt.plot(df2['Employed - Male'], label='Employed - Male', color='black')
plt.plot(df2['Employed - Female'], label='Employed - Female', color='red')
plt.legend(title='Group')
plt.ylabel('total employees and sub category of each state', fontsize=14)
plt.xlabel('states', fontsize=14)
```
